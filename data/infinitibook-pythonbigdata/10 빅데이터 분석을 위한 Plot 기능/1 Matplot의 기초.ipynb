## 직선 그래프

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt

y = [1, 2, 3, 4, 5]

plt.plot(y)
plt.show()
```

```python
y = [13, 16, 15, 18, 16, 17, 16]
plt.plot(y)
```

```python
plt.plot(range(len(y)), y);
```

```python
x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
plt.plot(x, y);
```

```python
plt.plot(x, y);
plt.xlabel('Daily')
plt.ylabel('No. of Book Sold');
```

```python
y1 = [13, 16, 15, 18, 16, 17, 16]
y2 = [17, 14, 17, 16, 15, 15, 14]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('Daily')
plt.ylabel('No. of Book Sold');
```

```python
plt.plot(x, y1, label='Sold')
plt.plot(x, y2, label='On Shelves')
plt.xlabel('Daily')
plt.ylabel('No. of Book Sold')
plt.legend(loc='upper left');
```

```python
plt.plot(x, y1, label='Sold')
plt.plot(x, y2, label='On Shelves')
plt.xlabel('Daily')
plt.ylabel('No. of Book Sold')
plt.legend(loc='upper left')
plt.title('Le Petit Prince: Sold & Left');
```

## 점선 그래프

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, 'bo')
plt.show()
```
