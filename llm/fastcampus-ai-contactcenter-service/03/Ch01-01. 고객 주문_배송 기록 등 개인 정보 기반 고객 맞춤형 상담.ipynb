```python
from string import Template

prompt_for_answer = Template("""
너는 사용자의 개인 정보 및 구매 기록에 기반하여 요청에 대한 답변을 생성해주는 인공지능 봇이야.

주어진 정보를 활용하여 사용자의 요청에 대해 친절하게 답변 해줘.

다음은 사용자의 요청과 요청에 관련된 정보야.

요청에 대해서 관련 정보 중 필요한 활용하여 도움이 되는 답변을 길고 친절하게 생성해 줘.

요청: $query

정보: $information

""")
```

```python
class CustomerDataBasedChatBot:
    def __init__(self, customer_data, log_data):
        self.customer_data = customer_data
        self.log_data = log_data
        print("사용자의 개인정보에 기반하여 답변을 생성해주는 챗봇입니다./n이름과 함께 질문을 해주세요./n구매하신 상품 정보도 같이 주신다면 더욱 정확하게 도와드릴 수 있습니다.")

    def query(self, query, name, product=None):
      customer_data = self.get_customer_data(name)
      log_data = self.get_log_data(name, product)

      information = customer_data + "/n" + log_data

      return information

    def get_customer_data(self, name):
      data = pandas_ai.run(self.customer_data, prompt=f'{name} 고객에 대한 정보를 가져와')
      return data

    def get_log_data(self, name, product=None):
      if product == None:
        data = pandas_ai.run(self.log_data, prompt=f'{name} 고객에 대한 정보를 가져와')
        return data
      else:
        data = pandas_ai.run(self.log_data, prompt=f'{name} 고객의 {product} 상품에 대한 정보만 자세하게 가져와')
        return data

customerDataBasedChatBot = CustomerDataBasedChatBot(customer_data, log_data)
```

```python
answer = customerDataBasedChatBot.query(query="상품 배송 상태에 대해 알 수 있을까요?", name="이철수", product="블랙 청바지")
```
