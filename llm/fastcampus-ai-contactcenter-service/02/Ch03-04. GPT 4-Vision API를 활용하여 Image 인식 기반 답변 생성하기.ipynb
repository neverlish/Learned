```python
def image_based_answer(image_url, query):
  chat_completion = client.chat.completions.create(
    model="gpt-4-vision-preview",
    messages=[
        {
              "role": "system",
              "content": "너는 강의 계획에 대한 다이어그램을 바탕으로 답변해 주는 인공지능 챗봇이야. 사용자의 질문에 대해 친절하게 답변해줘"
        },
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "query"},
                {
                    "type": "image_url",
                    "image_url": image_url
                }
            ]
        }
    ]
  )
  return chat_completion.choices[0].message.content

query_list = ["강의 계획에 음성 챗봇 관련 강의도 포함되어 있나요?", "금융 챗봇에 대해서는 무엇을 배울 수 있나요?"]
image_url = "https://storage.googleapis.com/static.fastcampus.co.kr/prod/uploads/202404/182557-1428/6.webp"

for q in query_list:
  print(f"Q: {q}")
  print(f"A: {image_based_answer(image_url,  q)}")
  print("\n===============================\n")
```

### 3. Multimodal Chain-of-Thought Prompting 을 사용하여 이미지에 대한 답변 생성하기
참고문헌: Language Is Not All You Need: Aligning Perception with Language Models (Huang et al., NeurIPS 2023, Microsoft) [[Paper]](https://arxiv.org/abs/2302.14045)

```python
def generate_image_details(image_url):
  chat_completion = client.chat.completions.create(
    model="gpt-4-vision-preview",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "해당 이미지에 대해 상세하게 설명해줘"},
                {
                    "type": "image_url",
                    "image_url": image_url
                }
            ]
        }
    ]
  )

  image_details = chat_completion.choices[0].message.content
  return image_details

def text_based_answer(details, query):
  system_prompt = "너는 강의 계획에 대한 설명을 바탕으로 답변해 주는 인공지능 챗봇이야. 사용자의 질문에 대해 친절하게 답변해줘."
  chat_completion = client.chat.completions.create(
      messages=[
          {
              "role": "system",
              "content": system_prompt
          },
          {
              "role": "user",
              "content": query,
          }
      ],
      model="gpt-3.5-turbo",
  )
  return chat_completion.choices[0].message.content

def multimodal_cot_answer(image_url, query):
  image_details = generate_image_details(image_url)
  answer = text_based_answer(image_details, query)
  return answer


query_list = ["강의 계획에 음성 챗봇 관련 강의도 포함되어 있나요?", "금융 챗봇에 대해서는 무엇을 배울 수 있나요?"]
image_url = "https://storage.googleapis.com/static.fastcampus.co.kr/prod/uploads/202404/182557-1428/6.webp"

for q in query_list:
  print(f"Q: {q}")
  print(f"A: {multimodal_cot_answer(image_url,  q)}")
  print("\n===============================\n")
```
