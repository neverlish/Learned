##### 음성 답변 챗봇 만들기

```python
model_name="gpt-3.5-turbo"
```

```python
from string import Template

query = "안녕하세요. 오늘 날짜로 뉴욕행 비행기를 예약할 수 있나요?"

system_prompt = """
너는 항공편 관련 상담을 해주는 인공지능 에이전트야.
"""

prompt = Template("""
다음 사용자의 요청에 대한 답변을 해줘

요청: $query
""")

print(system_prompt)
print(prompt.substitute(query=query))
```

```python
from IPython.display import Audio, display


class VoiceResponseChatbot:
  def __init__(self, system_prompt):
    self.system_prompt = system_prompt
    self.random_name = 1

  def query(self, question):
    chat_completion = client.chat.completions.create(
    messages=[
            {
                "role": "system",
                "content": system_prompt,
            },
            {
                "role": "user",
                "content": prompt.substitute(query=question),
            }
        ],
        model=model_name,
    )
    response_txt = chat_completion.choices[0].message.content

    self.tts(response_txt)

  def tts(self, txt):
    save_path = "/content/drive/MyDrive/데이터/ch0602"
    print(txt)
    speech_file_path = os.path.join(save_path, f"{self.random_name}.mp3")
    response = client.audio.speech.create(
      model="tts-1",
      voice="alloy",
      input=txt
    )

    response.stream_to_file(speech_file_path)

    display(Audio(speech_file_path, autoplay=True))

    self.random_name += 1

voiceResponseChatbot = VoiceResponseChatbot(system_prompt)
```

```python
voiceResponseChatbot.query("안녕하세요. 오늘 날짜로 뉴욕행 비행기를 예약할 수 있나요?")
```

```python
voiceResponseChatbot.query("성인 2명 비행기를 예약하고 싶어요")
```
