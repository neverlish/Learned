```python
import numpy as np

def get_embedding(text):
  return np.array(
    client.embeddings.create(
      model='text-embedding-ada-002',
      input=text,
      encoding_format='float'
    ).data[0].embedding
  )
```

```python
def cosine_similarity(vector_a, vector_b):
  dot_product = np.dot(vector_a, vector_b)

  norm_a = np.linalg.norm(vector_a)
  norm_b = np.linalg.norm(vector_b)

  cosine_similarity = dot_product / (norm_a * norm_b)

  return cosine_similarity
```

```python
result = cosine_similarity(get_embedding("kids"), get_embedding("children"))

print("Cosine Similarity: ", result)
```

```python
result = cosine_similarity(get_embedding("kids"), get_embedding("어린아이들"))

print("Cosine Similarity: ", result)
```

```python
def get_similarity(w1, w2):
  result = round(cosine_similarity(get_embedding(w1), get_embedding(w2)), 2)
  print(f"얼마나 가깝나요? {result}")

get_similarity("학교", "school")
get_similarity("bluetooth", "school")
```
