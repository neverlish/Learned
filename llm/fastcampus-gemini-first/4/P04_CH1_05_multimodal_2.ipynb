## 예시 돌려보기

```python
import base64
import vertexai
from vertexai.preview.generative_models import GenerativeModel, Part, Image
from IPython.display import Video

vertexai.init()
```

```python
def generate(video, text):
  model = GenerativeModel("gemini-1.5-flash")
  responses = model.generate_content(
    [video, text],
    generation_config={
        "max_output_tokens": 2048,
        "temperature": 0.4,
        "top_p": 1,
        "top_k": 32
    },
    stream=True,
  )
  
  for response in responses:
      print(response.text, end="")
```

```python
video1 = Part.from_data(data=base64.b64decode(""""""), mime_type="video/mp4")
```

```python
generate(video1, "이 영상 설명해줘")
```

## Local File 요청해보기

```python
Video("./data/sample.mp4", width=640)
```

```python
with open("./data/sample.mp4", "rb") as f:
    bin = f.read()
```

```python
video_encoded = base64.b64encode(bin)
```

```python
video2 = Part.from_data(data=video_encoded.decode('utf-8'), mime_type="video/mp4")
```

```python
generate(video2, "이 영상 설명해줘")
```
